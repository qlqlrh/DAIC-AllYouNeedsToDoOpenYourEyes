```python
import requests
from openai import OpenAI # openai==1.52.2
api_key = "up_qAeV4wltdGnqdmxavmqmxXyRBpmcZ" # ex: up_xxxYYYzzzAAAbbbCCC
file_path = "data/SPRI_AI_Brief_2023년12월호_F.pdf" # ex: ./image.png


client = OpenAI(
    api_key=api_key,
    base_url="https://api.upstage.ai/v1"
)

def chat_with_solar(messages):
    response = client.chat.completions.create(
        model="solar-pro",
        messages=messages
    )
    return response.choices[0].message['content']

# 대화 시작
# 모델이 대화 흐름을 이해할 수 있도록 이전 메시지들을 계속 messages 리스트에 누적시키는 방식입니다.
messages = [{"role": "user", "content": "Hello, who won the world series in 2020?"}]
response = chat_with_solar(messages)
print("Assistant:", response)
messages.append({"role": "assistant", "content": response})

# 다음 질문
# messages는 대화의 모든 흐름을 리스트로 누적합니다.
# 각 turn마다 user 의 질문과 → 모델의 답변인 assistant 형태로 계속 이어집니다.
messages.append({"role": "user", "content": "Where was it played?"})
response = chat_with_solar(messages)
print("Assistant:", response)
```

```python
from bs4 import BeautifulSoup

html = data['kwargs']['page_content']
soup = BeautifulSoup(html, 'html.parser')
paragraphs = []
charts = []


# 단락 텍스트 추출
for p in soup.find_all(['p']):
    category = p.get('data-category', 'paragraph')
    text = p.get_text(strip=True)
    paragraphs.append({'category': category, 'text': text})

# 도표(Figure/Chart) 추출
for figure in soup.find_all('figure'):
    fig_contents = figure.find('table')
    img_coords = figure.find('img').get('data-coord') if figure.find('img') else None
    if fig_contents and fig_contents.get_text(strip=True):
        charts.append({
            'contents': fig_contents.get_text(strip=True),
            'coordinates': img_coords,
        })
```
